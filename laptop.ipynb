##### Displaying total price of each model across different brands


```python
q3 = pd.read_sql_query("select company, product, sum(price_euros)as grand_total from laptop_price group by product order by grand_total desc ",db)
q3
```

##### Display all products by their company

```python
q4 = pd.read_sql_query("select company, product, typename, opsys from laptop_price order by company",db)
q4
```

##### Display total prices of product according to laptop type by company(Pivoting table)

```python
q5 = pd.read_sql_query("select distinct typename from laptop_price",db)
q5
```

```python
q6 = pd.read_sql_query("""select company, 
sum(case when typename='ultrabook' then price_euros else 0 end) as ultrabook,
sum(case when typename='notebook' then price_euros else 0 end) as notebook,
sum(case when typename='netbook' then price_euros else 0 end) as netbook,
sum(case when typename='gaming' then price_euros else 0 end) as gaming_laptop,
sum(case when typename='workstation' then price_euros else 0 end) as workstation_laptop,
sum(case when typename='2 in 1 convertible' then price_euros else 0 end) as convertible
from laptop_price 
group by company""",db)
q6 
```

##### Displaying highest price of each laptop type

```python
q7 = pd.read_sql_query("""select company, product, typename, Price_euros,
max(price_euros) over(partition by typename) as highest_price
from laptop_price order by highest_price asc""",db)
q7
```

##### Ranking the price of each brand by their price (rank() function)

```python
q8 = pd.read_sql_query("select * , rank() over(partition by company order by price_euros) as ranking from laptop_price",db)
q8
```

##### Displaying the top 2 highest price laptop from each brand(using CTE and WITH clause)


```python
q9 = pd.read_sql_query("""with rank_table as (select * , rank() over(partition by company order by price_euros desc) as ranking from laptop_price)
select * from rank_table rt where rt.ranking < 3""",db)
q9
```
